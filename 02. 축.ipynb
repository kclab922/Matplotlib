## 2. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # WIndows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시 마이너스 숫자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family': 'HYGungSo-Bold', 'size': 20}) # 개별 폰트 설정
```

```python
plt.plot(x, y)
plt.xlabel('X축')
plt.ylabel('y축')
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red')
plt.ylabel('y축', color='#00aa00') # RGB 사용 => RGB 중 G값만 설정
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red', loc='right') # left, center, right
plt.ylabel('y축', color='#00aa00', loc='top') # top, cener, bottom
```

```python
plt.plot(x,y)
plt.xticks([1, 2, 3]) # tick: 체크표시
plt.yticks([3, 6, 9, 12])
plt.show()
```

# 3. 범례 (legend)
그래프나 지도에서 표기된 특정 내용들을 설명(A legend is used to describe elements for a particular area of a graph.)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # WIndows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시 마이너스 숫자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, label='what data')
plt.legend()
plt.show()
```

```python
# 범례 위치 변경
plt.plot(x, y, label='what data')
plt.legend(loc='upper right')
```

```python
# 범례 위치 변경
plt.plot(x, y, label='what data')
plt.legend(loc='lower right') # loc 값은 공식문서 참고 
```

```python
# 범례의 세부위치 지정
plt.plot(x, y, label='범례')
plt.legend(loc=(0.7, 0.8)) # X축, Y축 기준 (0~1 사이)
```

# 8. 막대그래프 기본

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # WIndows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시 마이너스 숫자가 깨지는 현상을 해결
```

```python
labels = ['백호', '태웅', '대만']
values = [190, 187, 184]

plt.bar(labels, values)
```

```python
labels = ['백호', '태웅', '대만']
values = [190, 187, 184]

plt.bar(labels, values, color='r')
```

```python
labels = ['백호', '태웅', '대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors)
```

```python
labels = ['백호', '태웅', '대만']
values = [190, 187, 184]
colors = ['r', 'g']

plt.bar(labels, values, color=colors)
```

```python
labels = ['백호', '태웅', '대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha = 0.5)
```

```python
labels = ['백호', '태웅', '대만']
values = [190, 187, 184]

plt.bar(labels, values)
plt.ylim(175, 195) # y limit
```

```python
plt.bar(labels, values, width = 0.5)
```

```python
plt.bar(labels, values, width = 0.3)
plt.xticks(rotation=45) # X축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45)
```

```python
labels = ['백호', '태웅', '대만']
values = [190, 187, 184]
# ticks = ['1번', '2번', '3번']
ticks = [i for i in range(1, 4)]

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=90)
```

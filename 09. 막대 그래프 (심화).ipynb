```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # WIndows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시 마이너스 숫자가 깨지는 현상을 해결
```

```python
labels = ['백호', '태웅', '대만']
values = [190, 187, 184]

plt.barh(labels, values)
```

```python
labels = ['백호', '태웅', '대만']
values = [190, 187, 184]

plt.barh(labels, values)
plt.xlim(175, 195)
```

```python
# hatch : 창구
bar = plt.bar(labels, values)
bar[0].set_hatch('/') # //// 무늬 
bar[1].set_hatch('x') # 웨하스 무늬
bar[2].set_hatch('..') # 점 무늬
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for i, rect in enumerate(bar):
    plt.text(i, rect.get_height()+0.5, values[i], ha='center', color='b')
```
